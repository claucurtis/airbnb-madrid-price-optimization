```python
# Lectura y carga del set de datos 
df_host = pd.read_parquet("airbnb_madrid_host.parquet")
df_location = pd.read_parquet("airbnb_madrid_location.parquet")
df_property = pd.read_parquet("airbnb_madrid_property.parquet")
df_conditions = pd.read_parquet("airbnb_madrid_conditions.parquet")
df_reviews = pd.read_parquet("airbnb_madrid_reviews.parquet")
```

**3.1 Variables modificables:**
Dado que el objetivo del cliente es aumentaar el precio por noche de sus viviendas sin perder ocupación, se han elegido las siguientes variables en base a que sus caracteríticas son modificables, excepto una:
* Nº de camas
* Accomodates(capacidad máxima de huéspedes)
* Amenities 
* Condiciones de alquiler (política de cancelación)
* Valoraciones del alojamiento y host(nº reseñas, valoraciones medias, superhost)
* Nº de baños: variable no modificable inmediatamente.

**3.2 Hipótesis y preguntas de análsis (Roadmap):**
* H1: Nº camas, capacidad máxima de huéspedes y precio
    - ¿Aumentar el número de camas y una mayor capacidad máxima de huéspedes(accomodates) manteniendo el mismo número de habitaciones, presenta un incremento en el precio por noche?
    - Posibilidades: añadir sofá-camas, camas plegables, cambiar camas dobles por sencillas.

* H2: Amenities y precio
    - ¿Los apartamentos que están equipados con ciertos amenities presentan precios por noche más elevados que aquellos que no los tienen?
    - Posibilidades: Instalar A/C, tv, mejorar cocinas

* H3: Políticas de alquiler y precio
    - ¿Las diferentes condiciones de cancelación presentan diferencias en el precio por noche?
    - Posibilidades: Modificar las condiciones (hacia políticas más flexibles)

* H4: Puntuaciones medias del alojamiento y host
    - ¿Los apartamentos con mejores puntuaciones medias presentan precios más elevados por noche?
    - ¿Los apartamentos gestionados por Superhost cobran precios más elevados por noche?
    - Posibilidades: Mejorar experiencia del huésped y conseguir Superhost a la larga

        - Nota: para ser Superhost cumplir 4 requisitos en últimos 365 días:
            - 10+ reservas o 3 reservas con 100+ noches
            - tasa de respuesta >= 90% en menos de 24h
            - tasa cancelación inferior al 1%
            - valoración general >= 4.8 estrellas

* H5: Nº de baños (variable no modificable inmediatamente)
    - ¿Los apartamentos con más de un baño presentan precios por noche más elevados que los que solo tienen 1 baño?
    - Posibilidades: Plantearse hacer reformas en apartamentos con mayor viabilidad para aumentar el precio por noche y anunciarlo como alquiler de lujo.

H1.1: Nº camas

```python
# ¿Cuántas camas tienen los apartamentos?
# Algunos apartamentos tienen 0 camas(61) (camas mal registradas?, no registradas?)
df_cliente['beds'].value_counts().sort_index().head(8)
```

```python
# Se filtra el DF a apartamentos con > 0 camas para evitar errores de análisis
df_h1 = df_cliente[df_cliente['beds'] > 0]
print(f'Nº de apartamentos con camas registradas: {len(df_h1)}')
```

```python
# Precio mediano por número de camas
df_h1.groupby('beds')['price'].median().head(6)
```

```python
# Visulización nº camas vs precio/noche
sns.boxplot(data=df_h1[df_h1['beds'] <= 6], x='beds', y='price')
plt.title('Precio según nº de camas')
plt.xlabel('Nº de camas')
plt.ylabel('Precio (€/noche)')
plt.tight_layout()
plt.show()
```

```python
# Se observa que a más camas, mayor precio/noche. 
# Mayor diferencia entre 2 y 3 camas.
# Añadir 1 cama permitiría aumentar el precio 15€/noche 
precio_2_camas = df_h1[df_h1['beds']==2]['price'].median()
precio_3_camas = df_h1[df_h1['beds']==3]['price'].median()
diff_camas = precio_3_camas - precio_2_camas

print(f"Precio mediano 2 camas: {precio_2_camas:.2f}€")
print(f"Precio mediano 3 camas: {precio_3_camas:.2f}€")
print(f"Diferencia: {diff_camas:.2f}€")
```

```python
# Se observan mayores diferencias en precio mediano de 3 a 4 huéspedes máximos
# Aumentando la capacidad de 3 a 4 huéspedes en los mismos apartamentos
# se podría ganar 8,50€ más
precio_3p = df_h1[df_h1['accommodates']==3]['price'].median()
precio_4p = df_h1[df_h1['accommodates']==4]['price'].median()
diff_cap = precio_4p - precio_3p

print(f"Precio mediano 3 personas: {precio_3p:.2f}€")
print(f"Precio mediano 4 personas: {precio_4p:.2f}€")
print(f"Diferencia de 3 a 4 personas: {diff_cap:.2f}€")
```
